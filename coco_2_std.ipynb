```python
import cv2
from pathlib import Path
import json
from os.path import join, isdir, isfile
from os import makedirs
from matplotlib import pyplot as plt 
from modules import Bbox, ImageAnnotations

plt.rcParams['figure.figsize'] = [20, 15]
```

```python
images = [img.name for img in list(Path('images').glob('*.png'))]
js = json.load(open('annotations.json'))
annotations = js['annotations']
images_detail = {img_det['file_name']: img_det for img_det in js['images']}
img_id2img_name = {img_det['id']: img_det['file_name'] for img_det in js['images']}
```

```python
annots = {}

for img_id, img_path in img_id2img_name.items():
    temp_list = []
    for detail in annotations:
        if detail['image_id'] == img_id:
            bbox = detail['bbox']
            x1 = bbox[0]
            y1 = bbox[1]
            x2 = bbox[2]
            y2 = bbox[3]
            temp_list.append({
                'x1': x1, 
                'y1': y1, 
                'x2': x2, 
                'y2': y2,
                'category': 'shoe'
            
            }
            )
    annots[Path(img_path).name] = temp_list
    
    
```
